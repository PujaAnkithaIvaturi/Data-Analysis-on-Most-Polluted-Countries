```python
# Convert 'pollution_2023' to numeric
df['pollution_2023'] = pd.to_numeric(df['pollution_2023'], errors='coerce')

# Convert 'pollution_growth_Rate' to numeric
df['pollution_growth_Rate'] = pd.to_numeric(df['pollution_growth_Rate'], errors='coerce')

# Convert 'united_nation_Member' to boolean
df['united_nation_Member'] = df['united_nation_Member'].astype(bool)

# Convert 'share_borders' to string
df['share_borders'] = df['share_borders'].astype(str)

# Display basic information about the cleaned dataset
print("Basic Information about the Cleaned Dataset:\n", df.info())

# Display the cleaned DataFrame
print("Cleaned DataFrame:\n", df.head())
```

<b><u>Observation 1:</u> Assessing Pollution Levels by Country</b>

We sort the DataFrame based on the 'pollution_2023' column in descending order to identify the countries with the highest pollution levels.

We display the top 10 countries with the highest pollution levels and the top 10 countries with the lowest pollution levels.

We visualize the pollution levels for the top 20 countries using a bar chart.

```python
# Sort the DataFrame by pollution levels in descending order
sorted_df = df.sort_values(by='pollution_2023', ascending=False)

# Display the top 10 countries with the highest pollution levels
top_countries_highest_pollution = sorted_df[['country_name', 'pollution_2023']].head(10)
print("Top 10 Countries with the Highest Pollution Levels:\n", top_countries_highest_pollution)

# Display the top 10 countries with the lowest pollution levels
top_countries_lowest_pollution = sorted_df[['country_name', 'pollution_2023']].tail(10)
print("\nTop 10 Countries with the Lowest Pollution Levels:\n", top_countries_lowest_pollution)

# Visualize pollution levels using a bar chart
plt.figure(figsize=(12, 6))
plt.bar(sorted_df['country_name'][:20], sorted_df['pollution_2023'][:20], color='skyblue')
plt.title('Top 20 Countries with Highest Pollution Levels in 2023')
plt.xlabel('Country')
plt.ylabel('Pollution Level (2023)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize pollution levels using a bar chart
plt.figure(figsize=(12, 6))
plt.bar(sorted_df['country_name'][-20:][::-1], sorted_df['pollution_2023'][-20:][::-1], color='green')
plt.title('Top 20 Countries with Lowest Pollution Levels in 2023')
plt.xlabel('Country')
plt.ylabel('Pollution Level (2023)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Use Seaborn's Set3 color palette
colors = sns.color_palette("Set2")

# Group the data by continent and calculate the average pollution levels
average_pollution_by_continent = df.groupby('country_region')['pollution_2023'].mean().reset_index()

# Find the continent with the highest pollution level
max_pollution_continent = average_pollution_by_continent.loc[average_pollution_by_continent['pollution_2023'].idxmax()]

# Create a list to explode the pie chart
explode = [0.1 if region == max_pollution_continent['country_region'] else 0 for region in average_pollution_by_continent['country_region']]

# Visualize the comparison using a pie chart
plt.figure(figsize=(8, 6))
plt.pie(average_pollution_by_continent['pollution_2023'], labels=average_pollution_by_continent['country_region'],
        autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, wedgeprops=dict(edgecolor='black'))
plt.title('Average Pollution Levels by Continent in 2023')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```

```python
sns.set(style="whitegrid")

def plot_distribution(df, column, title, color="red"):
    """
    Function to plot both histogram and box plot for a given column in the dataset.
    """
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))

    sns.histplot(df[column], kde=True, ax=axes[0],  color="red")
    axes[0].set_title(f'Histogram of {title}')
    axes[0].set_xlabel(title)
    axes[0].set_ylabel('Frequency')

    sns.boxplot(x=df[column], ax=axes[1])
    axes[1].set_title(f'Box Plot of {title}')
    axes[1].set_xlabel(title)

    plt.tight_layout()
    plt.show()

# Plotting the distribution for 'pollution_growth_Rate'
plot_distribution(df, 'pollution_2023', 'Pollution_2023')

# Plotting the distribution for 'pollution_growth_Rate'
plot_distribution(df, 'pollution_growth_Rate', 'Pollution Growth Rate')

# Plotting the distribution for 'mostPollutedCountries_particlePollution'
plot_distribution(df, 'mostPollutedCountries_particlePollution', 'Particle Pollution Frequency')
```

```python
# Select relevant columns for analysis
correlation_columns = ['pollution_2023', 'country_land_Area_in_Km', 'pollution_density_in_km', 'pollution_density_per_Mile']

# Create a correlation matrix
correlation_matrix = df[correlation_columns].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Pollution and Related Factors')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of pollution levels using a histogram and box plot
def plot_distribution(df, column, title):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    
    sns.histplot(df[column], kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Histogram of {title}')
    axes[0].set_xlabel(title)
    axes[0].set_ylabel('Frequency')

    sns.boxplot(x=df[column], ax=axes[1], color='salmon')
    axes[1].set_title(f'Box Plot of {title}')
    axes[1].set_xlabel(title)

    plt.tight_layout()
    plt.show()

# Plot the distribution of pollution levels
plot_distribution(df, 'pollution_2023', 'Pollution 2023')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a new column indicating whether the country shares borders or not
df['shares_borders'] = df['share_borders'].apply(lambda x: 'Shares Borders' if x else 'Does Not Share Borders')

# Create a box plot to compare pollution levels for countries that share borders and those that do not
plt.figure(figsize=(12, 8))
sns.boxplot(x='shares_borders', y='pollution_2023', data=df, palette='Set2')

# Add labels and title
plt.xlabel('Border Sharing')
plt.ylabel('Pollution in 2023')
plt.title('Impact of Border Sharing on Pollution Levels in 2023')

# Display the plot
plt.show()
```

```python
#!pip install geopandas
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a world map shapefile
world_map = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world map with your pollution data
merged_data = world_map.merge(df, how='left', left_on='name', right_on='country_name')

# Plot the choropleth map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='pollution_2023', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add a title and legend
plt.title('Pollution Levels Across Countries in 2023')
ax.set_axis_off()
plt.show()
```
